# Tiền xử lý dữ liệu
## 1. Làm sạch dữ liệu
### 1.1. Kiểm tra dữ liệu bị thiếu

```python
print(df.isnull().sum())
```

### 1.2. Phát hiện dữ liệu nhiễu

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Vẽ boxplot để kiểm tra outliers
features = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']
plt.figure(figsize=(12, 8))
for i, col in enumerate(features):
    plt.subplot(2, 3, i+1)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

### 1.3. Biến đổi dữ liệu (Log Transform và chuẩn hóa)

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

# Log transform để giảm skewness
X_log = np.log1p(df[features]) #np.log1p(): xử lý các giá trị lớn bất thường

# Chuẩn hóa dữ liệu z-score
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_log)
```

### 1.4. Giảm chiều dữ liệu (dùng PCA để trực quan hóa)

```python
from sklearn.decomposition import PCA

# Giảm xuống 2 chiều để trực quan hóa
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Vẽ dữ liệu sau PCA
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], alpha=0.5)
plt.title("Dữ liệu sau PCA (2 chiều)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

# Thuật toán DBSCAN
## 1. Chọn tham số epsilon bằng phương pháp khuỷu tay
min_samples thường chọn bằng 2 × số chiều dữ liệu (6 chiều → min_samples = 12). Do hai thuộc tính Channel và Region là hai thuộc tính có phân phối rời rạc như hình đã vẽ ở bước trên, nên chúng ta loại bỏ hẳn 2 thuộc tính này, không đưa vào tính toán. Ví dụ, thuộc tính Region sẽ có các giá trị {1,2,3}, thuật toán thì coi 3 là xa gấp 3 lần so với 1,2. Điều này dẫn đến nhiều sai lệch khi phân cụm.

```python
from sklearn.neighbors import NearestNeighbors
# Tìm khoảng cách đến k-th nearest neighbor (ở đây k = 4)
k = 4  
neighbors = NearestNeighbors(n_neighbors=k)
neighbors_fit = neighbors.fit(X_scaled)
distances, indices = neighbors_fit.kneighbors(X_scaled)

# Sắp xếp khoảng cách lớn nhất của mỗi điểm
sorted_distances = np.sort(distances[:, k-1])
plt.plot(sorted_distances)
plt.xlabel("Điểm dữ liệu đã sắp xếp")
plt.ylabel(f"Khoảng cách đến {k} điểm gần nhất")
plt.title("Phương pháp khuỷu tay cho DBSCAN")
plt.show()
```

```python

from kneed import KneeLocator

# Tìm điểm khuỷu tay
i = np.arange(len(sorted_distances))
knee = KneeLocator(i, sorted_distances, S=1, curve='convex', direction='increasing', interp_method='polynomial')

# Vẽ biểu đồ và in ra epsilon
plt.figure(figsize=(6, 4))
knee.plot_knee()
plt.xlabel("Điểm dữ liệu đã sắp xếp")
plt.ylabel(f"Khoảng cách đến {k} điểm gần nhất")
plt.title("Phương pháp khuỷu tay cho DBSCAN (KneeLocator)")
plt.show()

print("Epsilon tối ưu =", sorted_distances[knee.knee])
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

# Áp dụng DBSCAN
dbscan = DBSCAN(eps=1.5, min_samples=4)
labels = dbscan.fit_predict(X_scaled)

# Gắn nhãn vào dataframe
df['Cluster'] = labels

# Kiểm tra kết quả
print("Số lượng cụm:", len(set(labels)) - (1 if -1 in labels else 0))
print("Số lượng nhiễu:", list(labels).count(-1))

# Silhouette Score (nếu có >1 cụm)
if len(set(labels)) > 1 and -1 not in set(labels):
    score = silhouette_score(X_scaled, labels)
    print("Silhouette Score:", score)

# Trực quan hóa cụm
plt.figure(figsize=(8,6))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=labels, palette='Set1')
plt.title("Phân cụm DBSCAN sau PCA")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score
import numpy as np
import pandas as pd

# Danh sách các giá trị epsilon cần thử
eps_values = np.arange(1.2, 3.0, 0.2)  # từ 1.2 đến 2.5, bước 0.2
minPts = 12

# Lưu kết quả vào danh sách
results = []

for eps in eps_values:
    dbscan = DBSCAN(eps=eps, min_samples=minPts)
    labels = dbscan.fit_predict(X_scaled)
    
    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise = list(labels).count(-1)
    
    # Kiểm tra điều kiện để tính các chỉ số
    if n_clusters >= 1 and n_noise < len(labels):
        sil_score = silhouette_score(X_scaled, labels)
        db_index = davies_bouldin_score(X_scaled, labels)
    else:
        sil_score = None
        db_index = None
    
    results.append({
        'epsilon': eps,
        'clusters': n_clusters,
        'noise_points': n_noise,
        'silhouette_score': sil_score,
        'davies_bouldin_index': db_index
    })

# Hiển thị kết quả dưới dạng bảng
results_df = pd.DataFrame(results)
print(results_df)

```

```python
import matplotlib.pyplot as plt

# Loại bỏ các dòng có giá trị None
filtered_df = results_df.dropna(subset=['silhouette_score', 'davies_bouldin_index'])

# Vẽ biểu đồ
plt.figure(figsize=(12,5))

# Silhouette Score
plt.subplot(1,2,1)
plt.plot(filtered_df['epsilon'], filtered_df['silhouette_score'], marker='o', color='green')
plt.title("Silhouette Score theo epsilon")
plt.xlabel("epsilon")
plt.ylabel("Silhouette Score")
plt.grid(True)

# Davies-Bouldin Index
plt.subplot(1,2,2)
plt.plot(filtered_df['epsilon'], filtered_df['davies_bouldin_index'], marker='o', color='red')
plt.title("Davies-Bouldin Index theo epsilon")
plt.xlabel("epsilon")
plt.ylabel("Davies-Bouldin Index")
plt.grid(True)

plt.tight_layout()
plt.show()
```
